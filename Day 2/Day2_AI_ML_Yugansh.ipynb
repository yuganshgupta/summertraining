# WAP TO create a auto threshold algorithm
1. Create a 9x9 matrix within a range of 0-100 intger number
2.find the **mean** of each row and save it in a B list **(1x9)**
2.step 2= [B0,B1,B2,B3,B4,B5,B6,B7,B8]
3. NOW,COMPARE the elements of each row with its respective mean , if mean is greater than the element than the element , then leave the element as it . else make the element 0.create  a new matrix C to hold the updated result
4.Count of the number of zeros and non zeros elements , which should be equal to the number of number of elements in a matrix

```python

np.random.seed(24242)
mat1 = np.random.randint(0,100,size=(9,9))
print("The old matrix formed\n")
print(mat1)


#initialize the value of the list for storing the mean values
mean_val= np.zeros(mat1.shape[0])
no_of_elements= 0


#Running the loop to calculate mean and store each value of mean
for i in range(0,mat1.shape[0]):
  m = np.mean(mat1[i,:])
  mean_val[i] = m

print("The mean values stored\n")
print(mean_val)
#Initializing the counter for zeros and non zeros number
Nonzeros_count=0
Zeros_count=0



mat2= np.zeros((mat1.shape[0],mat1.shape[1]))
#Running the for loop to implement step 3
for i in range(mat1.shape[0]):
  for j in range(mat1.shape[1]):
    if mat1[i,j] > mean_val[i]:
      mat2[i,j] = mat1[i,j]
      Nonzeros_count+=1
    else:
      mat2[i,j] = 0
      Zeros_count+=1


#Printing the new matrix that has been formed
print("The new matrix formed\n")
print(mat2)


print()
#printing the values of non zeros and zeros elements
print("The number of non zeros elements are ",Nonzeros_count)
print("The number of zeros elements are ",Zeros_count)

```

```python
import numpy as np

#Creating a matrix for 4x4
np.random.seed(2442)
nat1 = np.random.randint(0,30,size=(4,4))
print(f"{nat1}\n\n")

#Creating the window that is half the size of the initial matrix
n1s=int(nat1.shape[0]/2)
n2s=int(nat1.shape[1]/2)
window= np.zeros((n1s,n2s))
print(f"\n\n{window}\n\n")

# Initialize the result matrix
res=np.zeros((n1s,n2s))
print(f"\n\n{window}\n\n")

for i in range(0,nat1.shape[0],2):
    for j in range(0,nat1.shape[1],2):
        # Extract the 2x2 window
        B = nat1[i:i+2,j:j+2]
        # Compute the max value in the window
        con=np.max(B)
        # Store the max value in the corresponding cell in the result matrix
        res[i//2, j//2] = con
        print(f"{B}\n\n")

# Print the result matrix
print(res)
```
