```python
neigh_mat=np.zeros((num_of_sensors,num_of_sensors))
neigh_mat
```

```python
SR=30 # SR = Sensor range
for i in range(dist_mat.shape[0]):
  for j in range(dist_mat.shape[0]):
    if(i!=j):
      if(dist_mat[i,j]<SR):
        neigh_mat[i,j]=1
    else:
       neigh_mat[i,j]=0
```

```python
import numpy as np
import matplotlib.pyplot as plt

num = 100
raw_data = 0+100*np.random.random((num,2))
c1x = raw_data[0,0];c1y = raw_data[0,1]
c2x = raw_data[1,0];c2y = raw_data[1,1]
c3x = raw_data[2,0];c3y = raw_data[2,1]
c4x = raw_data[3,0];c4y = raw_data[3,1]

plt.figure(1,(8,8))
plt.plot(raw_data[:,0],raw_data[:,1],'ro')

for it in range(10):
    cluster1x = []; cluster1y = []
    cluster2x = []; cluster2y = []
    cluster3x = []; cluster3y = []
    cluster4x = []; cluster4y = []

    for i in range(raw_data.shape[0]):
        d1=np.sqrt((c1x-raw_data[i,0])**2+(c1y-raw_data[i,1])**2)
        d2=np.sqrt((c2x-raw_data[i,0])**2+(c2y-raw_data[i,1])**2)
        d3=np.sqrt((c3x-raw_data[i,0])**2+(c3y-raw_data[i,1])**2)
        d4=np.sqrt((c4x-raw_data[i,0])**2+(c4y-raw_data[i,1])**2)

        if(d1<d2 and d1<d3 and d1<d4):
            cluster1x.append(raw_data[i,0])
            cluster1y.append(raw_data[i,1])
            plt.plot(raw_data[i,0],raw_data[i,1],'bo')

        elif(d2<d1 and d2<d3 and d2<d4):
            cluster2x.append(raw_data[i,0])
            cluster2y.append(raw_data[i,1])
            plt.plot(raw_data[i,0],raw_data[i,1],'go')

        elif(d3<d1 and d3<d2 and d3<d4):
            cluster3x.append(raw_data[i,0])
            cluster3y.append(raw_data[i,1])
            plt.plot(raw_data[i,0],raw_data[i,1],'ro')

        else:
            cluster4x.append(raw_data[i,0])
            cluster4y.append(raw_data[i,1])
            plt.plot(raw_data[i,0],raw_data[i,1],'co')

    c1x = sum(cluster1x)/len(cluster1x);c1y = sum(cluster1y)/len(cluster1y)
    c2x = sum(cluster2x)/len(cluster2x);c2y = sum(cluster2y)/len(cluster2y)
    c3x = sum(cluster3x)/len(cluster3x);c3y = sum(cluster3y)/len(cluster3y)
    c4x = sum(cluster4x)/len(cluster4x);c4y = sum(cluster4y)/len(cluster4y)

    plt.title(["Iteration number: "+str(it)])
    plt.pause(0.5)
```
