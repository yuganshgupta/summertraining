```python
# WAP to perform arithmetic operations +-*/ on the two numbers
# entered by user and operation operator is also entered by the user which means entered by the user
# Do you want to continue if y (repeat) and if n ( end  it )
# with proper message


flag = True
while flag:
  n1=int(input("Enter the first number "))
  n2=int(input("Enter the second number "))
  operation=input("Enter the operation you want to perform ")

  if operation == "+":

    print(f"The sum of {n1} and {n2} is {n1+n2}")

  if operation == "-":

    print(f"The difference of {n1} and {n2} is {n1-n2}")

  if operation == "*":

    print(f"The product of {n1} and {n2} is {n1*n2}")

  if operation == "/":

    print(f"The division  of {n1} and {n2} is {n1/n2}")

  choice=input("Do you want to continue yes : y or n :")
  if choice == "y":
    flag = True
  else:
    flag = False
```

```python
def user_input():
  n1=int(input("Enter the first number "))
  n2=int(input("Enter the second number "))
  operation=input("Enter the operation you want to perform ")
  return n1,n2,operation

def custom_operation(n1,n2,operation):
    if operation == "+":

      print(f"The sum of {n1} and {n2} is {n1+n2}")

    if operation == "-":

      print(f"The difference of {n1} and {n2} is {n1-n2}")

    if operation == "*":

      print(f"The product of {n1} and {n2} is {n1*n2}")

    if operation == "/":

      print(f"The division  of {n1} and {n2} is {n1/n2}")

def user_choice():
  choice=input("Do you want to continue yes : y or n :")
  if choice == "y":
    flag = True
    return flag
  else:
    flag = False
    return flag

flag = True
while flag:
  n1,n2,operation=user_input()
  custom_operation(n1,n2,operation)
  flag= user_choice()




```

```python
import matplotlib.pyplot as plt
import numpy as np
# sin function
# cos function
# custom function
X = np.linspace(1,10,100)
Y = np.sin(X)
Z = np.cos(X)
W = 2*np.sin(X)+np.cos(X)
P = np.sin(X) - 2*np.cos(X)
```

```python
# all functions output in different figure
plt.figure(1,figsize=(5,2))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("sine function")
plt.xlabel('Range')
plt.ylabel('sin(X)')
plt.grid('on')

plt.figure(2,figsize=(5,2))#activting the figure number 1
plt.plot(X,Z,'g',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("cos function")
plt.xlabel('Range')
plt.ylabel('cos(X)')
plt.grid('on')

plt.figure(3,figsize=(5,2))#activting the figure number 1
plt.plot(X,W,'b',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')

plt.figure(4,figsize=(5,2))#activting the figure number 1
plt.plot(X,P,'k',linewidth=4) # 'r','g,'m','c','k','y','b'
plt.title("custom function")
plt.xlabel('Range')
plt.ylabel('func(X)')
plt.grid('on')
```

```python
# comparative analysis
plt.figure(1,figsize=(10,4))#activting the figure number 1
plt.plot(X,Y,'r',linewidth=4,label="sin(X)") # 'r','g,'m','c','k','y','b'
plt.plot(X,Z,'g',linewidth=4,label='Cos(X)')
plt.plot(X,W,'b',linewidth=4,label='func1(X)')
plt.plot(X,P,'m',linewidth=4,label='func2(X)')
plt.title("comparative analysis")
plt.xlabel('Range')
plt.ylabel('function(X)')
plt.grid('on')
plt.legend() # activate the legends
```
