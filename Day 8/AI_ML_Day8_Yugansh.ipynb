```python
data = np.random.randint(0,100,(500,2))

print(data.shape)

# visualize the dummy data
plt.figure(1,(4,4))
plt.scatter(data[:,0],data[:,1])
plt.title('Raw data')
```

```python
km_model = cluster.KMeans(n_clusters = 2, random_state= 52)
km_model = km_model.fit(data)
print(km_model.cluster_centers_)
```

```python
# visualize the dummy data
plt.figure(1,(8,4))
plt.subplot(1,2,1)
plt.scatter(data[:,0],data[:,1])
plt.title('Raw data')
plt.subplot(1,2,2)
plt.scatter(data[:,0],data[:,1],c=km_model.labels_)
plt.title('clustered data')
for i in range(len(km_model.cluster_centers_)):
    dx = km_model.cluster_centers_[i][0]
    dy = km_model.cluster_centers_[i][1]
    plt.plot(dx,dy,'bd')
```

```python
# Import necessary libraries
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Define the path to the image and the Haar cascade XML file
# Note: You need to upload your image file to Google Colab and adjust the path accordingly
image_path = '/content/image_faces.png'
# image_path = '/content/h1.jpg'
# image_path = '/content/h2.jpg'
cascade_path = cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'

# Read the image
im = cv2.imread(image_path)

# Check if the image was loaded successfully
if im is None:
    print("Error loading image")
else:
    # Resize the image
    im_new = cv2.resize(im, (512,512))

    # Print the resolution of the image
    print("Resolution of the image")
    print(im.shape)

    # Convert the color (BGR) into grayscale
    gray_im = cv2.cvtColor(im_new, cv2.COLOR_BGR2GRAY)

    # Print the resolution of the gray image
    print("Resolution of the gray image")
    print(gray_im.shape)

    # Initialize the face detector
    face_detector = cv2.CascadeClassifier(cascade_path)

    # Run the classifier on the image
    faces = face_detector.detectMultiScale(gray_im, scaleFactor=1.1, minNeighbors=10)

    # Print the detected faces
    print(faces)

    # Display the bounding box on all the faces
    for (dx, dy, w, h) in faces:
        cv2.rectangle(im_new, (dx, dy), (dx+w, dy+h), (255, 0, 0), 2)

    # Display the image with detected faces
    cv2_imshow(im_new)
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time
from google.colab.patches import cv2_imshow
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode
import io
from PIL import Image

# Function to capture video from the webcam
def capture_video():
    js = Javascript('''
    async function captureVideo() {
      const video = document.createElement('video');
      video.style.display = 'none';
      document.body.appendChild(video);

      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      video.srcObject = stream;
      await new Promise((resolve) => video.onloadedmetadata = resolve);

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      const context = canvas.getContext('2d');
      context.drawImage(video, 0, 0, canvas.width, canvas.height);
      stream.getTracks().forEach(track => track.stop());
      document.body.removeChild(video);
      return canvas.toDataURL('image/jpeg');
    }
    captureVideo();
    ''')
    display(js)
    data = eval_js('captureVideo()')
    binary = b64decode(data.split(',')[1])
    image = Image.open(io.BytesIO(binary))
    return np.array(image)

# Load the Haar Cascade file
path = "/content/haarcascade_frontalface_default.xml"  # Update this path
face_detector = cv2.CascadeClassifier(path)

frame = True
while frame:
    im = capture_video()
    im_new = cv2.resize(im, (512, 512))
    gray_im = cv2.cvtColor(im_new, cv2.COLOR_BGR2GRAY)
    faces = face_detector.detectMultiScale(gray_im, scaleFactor=1.1, minNeighbors=10)

    for (dx, dy, w, h) in faces:
        cv2.rectangle(im_new, (dx, dy), (dx + w, dy + h), (255, 0, 0), 2)

    cv2_imshow(im_new)

    if cv2.waitKey(1) & 0xFF == ord('q'):
        frame = False
        break

cv2.destroyAllWindows()
```
